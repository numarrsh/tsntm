```python
def prepare_instances(data_df):
    instances = []
    for idx_doc, doc in data_df.iterrows():
        instance = Instance()
        instance.idx = idx_doc
        instance.bow = doc.doc_bow
        instances.append(instance)
    return instances
```

```python
instances_train = prepare_instances(train_df)
instances_dev = prepare_instances(dev_df)
instances_test = prepare_instances(test_df)
```

```python
path_output = '/home/m-isonuma/tsntm/data/yelp/instances.pkl'
cPickle.dump((instances_train, instances_dev, instances_test, word_to_idx, idx_to_word, bow_idxs),open(path_output,'wb'))
```
